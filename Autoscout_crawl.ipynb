```python
listCars =[]
```

```python
for carNr in dict_details_data:
    dictCars1 = {}
    dictCars1["ID"]= carNr
    soup = BeautifulSoup(dict_details_data[carNr],"html.parser")
    for car in soup.find("div", class_="cldt-price").stripped_strings:
        dictCars1["Preis"] = int(car.strip("€,.- ").replace(".", ""))
    for car in soup.find("span", class_="sc-font-l cldt-stage-primary-keyfact").stripped_strings:
        dictCars1["Laufleistung"] = int(car.strip("€,.- km").replace(".", ""))
    listCars.append(dictCars1)
```
